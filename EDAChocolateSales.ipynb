Import libraries & load dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Chocolate_Sales.csv")
df.head()
```

Check data structure

```python
df.shape
df.columns
df.info()
df.describe()
```

The dataset contains sales information about chocolate products including country, product, Amount, Boxes Shipped, and date.

Missing values

```python
df.isnull().sum()
df.dropna(inplace=True)
```

Remove duplicates

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

Data Visualization

Top selling products

```python
df['Product'].value_counts().plot(kind='bar', figsize=(8,5))
plt.title("Top Selling Chocolate Products")
plt.show()
```

Sales by country

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Country', data=df)
plt.xticks(rotation=45)
plt.title("Sales by Country")
plt.show()
```

Amount Distribution

```python
sns.histplot(df['Amount'], kde=True)
plt.title("Amount Distribution")
plt.show()
```

Relationship Boxes Shipped & Amount


```python
sns.scatterplot(x='Boxes Shipped', y='Amount', data=df)
plt.title("Boxes Shipped & Amount")
plt.show()
```
